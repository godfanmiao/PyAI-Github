```python
import pandas as pd
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'V': ['V1', 'V2', 'V3', 'V4']})

right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                      'U': ['U0', 'U1', 'U2', 'U3']})

#将两个数据框，依照指定的列，采取行合并。
pd.merge(left, right, on='key')
```

```python
#将两个数据框，通过行对齐的方式进行合并。
pd.concat([left, right], axis=1)
```

```python
#将两个数据框，通过列对齐的方式进行合并。
pd.concat([left, right], axis=0)
```
